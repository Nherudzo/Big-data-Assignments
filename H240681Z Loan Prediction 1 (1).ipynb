```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_auc_score
```

```python
LoanPred = pd.read_csv(r"C:/Users/NHEE/Downloads/loan prediction based on customer behavior dataset.csv")
```

```python
LoanPred.head()
```

```python
LoanPred.isnull()
```

```python
LoanPred.dtypes
```

```python
LoanPred2 = LoanPred.drop(columns = ["Id","CITY" ,"STATE","Profession"])
```

```python
LoanPred2.head()
```

```python
LoanPred2["Married/Single"].replace({"married":0,"single":1},inplace=True)
```

```python
LoanPred2.head()
```

```python
for col in ["House_Ownership","Car_Ownership"]:
    LoanPred2[col] = LabelEncoder().fit_transform(LoanPred2[col])
    
```

```python
LoanPred2.head()
```

```python
x = LoanPred2.drop("Risk_Flag",axis = 1)
y = LoanPred2["Risk_Flag"]
                   
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
RF= RandomForestClassifier()
RF.fit(x_train,y_train)
LogisticReg = LogisticRegression()
LogisticReg.fit(x_train,y_train)
```

```python
# Model Prediction and Evaluation
pred1 = LogisticReg.predict(x_test)
pred2 = RF.predict(x_test)
print(classification_report(y_test,pred1))
print(classification_report(y_test,pred2))

      
```

```python
#The Random Forest model is better due to its ability to predict both classes more effectively, especially the minority class, while maintaining high overall perfomance. The Logistic Regression model is inadequate for this task as it fails to identify the minority class entirely.
```

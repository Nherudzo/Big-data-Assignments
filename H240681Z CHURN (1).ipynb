```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_auc_score
```

```python
CHURN = pd.read_csv(r"C:\Users/NHEE/Downloads/Bank Customer Churn Prediction.csv")
```

```python
for col in ["country","gender"]:
    CHURN[col]= LabelEncoder().fit_transform(CHURN[col])
```

```python
CHURN2 = CHURN.drop(columns = ["customer_id"])
```

```python
CHURN2.head()
```

```python
x = CHURN2.drop("churn",axis = 1)
y = CHURN2["churn"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
RF= RandomForestClassifier()
RF.fit(x_train,y_train)
LogisticReg = LogisticRegression()
LogisticReg.fit(x_train,y_train)
```

```python
CHURN1 = LogisticReg.predict(x_test)
CHURN2 = RF.predict(x_test)
print(classification_report(y_test,CHURN1))
print(classification_report(y_test,CHURN2))
```

```python
#The Random Forest model performs significantly better than the Logistic Regression model  for this churn prediction task. While both models excel at identifying non-churn instances, Random Forest demonstrates a much stronger ability to detect churn instances, as evidenced by improved precision, recall, and F1-Score for Class 1. The overall accuracy and weighted averages also favor Random Forest.Random Forest outperforms Logistic Regression  for churn prediction
```
